```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
raw_lists = pd.read_csv("raw/listings.csv")
print(raw_lists.columns)
```

How many hosts are there ?

```python
len(raw_lists["host_id"].unique())
```

What about multiple listings`?

```python
raw_lists["host_id"].value_counts()
```

```python
plt.hist(raw_lists["host_id"].value_counts().reset_index()["host_id"],bins=40)
plt.show()
plt.boxplot(raw_lists["host_id"].value_counts().reset_index()["host_id"],vert=False)
plt.xlabel("No. of Listings")
plt.show()
```

```python
listings_summary = pd.read_csv("listings.csv")
listings_summary.groupby(["host_id","host_name","calculated_host_listings_count"]).mean()["price"].reset_index().sort_values(by=["calculated_host_listings_count"],ascending=False).head(50)
```
